

```python if,elif,else

    statements
elif(logic):
    statements

```

````python
# write a program to check given number is even number or odd number
num=int(input("please enter first number here:"))
if(num%2==0):
    print(num,"is even number")
else:
    print(num,"is odd number")
````

````python
# write a program to find the number is negative or positive or zero
n=int(input("enter your first number here:"))
if (n>0):
    print("positive number")
elif(n<0):
    print("negative number")
else:
    print("zero")
````
